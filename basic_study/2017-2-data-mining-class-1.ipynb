```r
#install.packages('tm')
#install.packages('SnowballC')
#install.packages('wordcloud')
#install.packages('topicmodels')

library(tm)
library(SnowballC)
library(wordcloud)
library(topicmodels)
```

```r
jeopardy<-read.csv("jeopardy.csv",stringsAsFactors = FALSE, nrows=1000)

data_corpus <- Corpus(VectorSource(jeopardy$Question)) # 말뭉치
data_corpus <- tm_map(data_corpus, content_transformer(tolower)) #소문자
data_corpus <- tm_map(data_corpus, removePunctuation) # 구두점 제거
data_corpus <- tm_map(data_corpus, removeNumbers) # 숫자제거
data_corpus <- tm_map(data_corpus, removeWords, stopwords('english')) # 불용어 제거
data_corpus <- tm_map(data_corpus, stemDocument) # 어근(단어원형) 추출
```

```r

wordcloud(data_corpus, max.words=100, random.order=F)

TDM <- TermDocumentMatrix(data_corpus, control = list(minWordLength = 1))

inspect(TDM[1:10, 1:10])

TFIDF <- weightTfIdf(TDM, normalize=FALSE)
inspect(TFIDF[1:10, 1:10])

findFreqTerms(TDM, lowfreq=10)

findAssocs(TDM, "type", 0.2) # t(TDM)을 이용하여 문서간 상관관계높은 문서도 추출 가능



```

```r
#install.packages('arules')
library(arules)
BIN <- weightBin (TDM)
BIN_mat <- as.matrix(BIN)
BIN_trans <- as(t(BIN_mat), "transactions")
rules <-apriori(BIN_trans, parameter=list(support=0.005, confidence=0.1, minlen=2))
inspect(sort(rules, by = "lift")[1:20])
```

```r
# SVD decomposition
SVD <- svd(TDM)
LSI.U <- SVD$u
LSI.D <- SVD$d
LSI.V <- SVD$v
# Documents Mapping
Document.Mat <- t((diag(2)*LSI.D[1:2]) %*% t(LSI.V[,1:2]))
plot(Document.Mat[,1], Document.Mat[,2], xlab = "first SV", ylab = "second SV", main
     = "Documents mapping")
text(Document.Mat[1:10,1], Document.Mat[1:10,2])
# Terms Mapping
Term.Mat <- LSI.U[,1:2] %*% (diag(2)*LSI.D[1:2])
plot(Term.Mat[,1], Term.Mat[,2], xlab = "first SV", ylab = "second SV", main = "Terms
mapping")
text(Term.Mat[1:10,1], Term.Mat[1:10,2])
```

```r


#install.packages("topicmodels")
library(topicmodels)

DTM <- DocumentTermMatrix(data_corpus, control = list(minWordLength = 1))

rowTotals <- apply(DTM , 1, sum) #Find the sum of words in each Document
DTM.new <- DTM[rowTotals> 0, ] #remove all docs without words
lda.output <-LDA(DTM.new, control=list(seed=10), k=5)
dim(lda.output@gamma) # 문서 x 토픽 행렬 크기
dim(lda.output@beta) # 토픽 x 단어 행렬 크기

terms(lda.output, 12)

posterior_lda <- posterior(lda.output)

round(posterior_lda$topics, 3)

round(posterior_lda$terms, 3)[, 1:10] 

```
